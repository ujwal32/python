# Relu

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

# Create some input values
x = np.linspace(-10, 10, 100)
y = relu(x)

# Plotting the ReLU function
plt.plot(x, y, label='ReLU(x)', color='blue')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

inputs = np.array([-5, -1, 0, 2, 5])
outputs = relu(inputs)

print("Input: ", inputs)
print("ReLU Output: ", outputs)
```

# softmax

```python
import numpy as np

def softmax(x):
    exps = np.exp(x - np.max(x))  # for numerical stability
    return exps / np.sum(exps)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exps = np.exp(x - np.max(x))  # stability trick
    return exps / np.sum(exps)

# Generate a range of x values
x_vals = np.linspace(-5, 5, 100)

# Store probabilities for 3 classes
class_1 = []
class_2 = []
class_3 = []

for x in x_vals:
    logits = np.array([x, 1, 0])  # change only first value
    probs = softmax(logits)
    class_1.append(probs[0])
    class_2.append(probs[1])
    class_3.append(probs[2])

# Plotting
plt.plot(x_vals, class_1, label='Class 1 (x)', color='blue')
plt.plot(x_vals, class_2, label='Class 2 (1)', color='green')
plt.plot(x_vals, class_3, label='Class 3 (0)', color='red')
plt.title('Softmax Probability Distribution')
plt.xlabel('Logit Value for Class 1')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```
